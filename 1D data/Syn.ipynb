```python
import pandas as pd
import numpy as np
import random

# Constants
rows = 50
p0_z_dimensions = 11
stored_soln_dimensions = 72
filename_base = "SSE_scenario_Simple_"
assignments = ["training", "validation", "test"]
data = {
    "p0_z": [np.random.rand(p0_z_dimensions).tolist() for _ in range(rows)],
    "Stored_Soln": [np.random.rand(stored_soln_dimensions).tolist() for _ in range(rows)],
    "Filename": [f"{filename_base}{random.randint(1,100)}.mat" for _ in range(rows)],
    "assignment": [random.choice(assignments) for _ in range(rows)]
}


df = pd.DataFrame(data)
df.to_pickle("generated_data.pkl")
df
```
